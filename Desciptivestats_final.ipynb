```
import pandas as pd
data = pd.read_csv("student-mat.csv", sep = ';') 
data
```

```
mean_students = data.mean()
mean_students
```

```
from scipy import stats
gm = ["traveltime","freetime","studytime"] #geometric mean
 
gmean_time = stats.gmean(data.loc[:,gm])
gmean_time
```

```
hmean_time = stats.hmean(data.loc[:,gm])  #harmonic mean
hmean_time
```

```
details_students = data.describe() #mean, standard deviation, min, max,quartiles
details_students
```

```
predata_interquartile_25 = details_students.iloc[4]  #interquartile range
predata_interquartile_25
predata_interquartile_75 = details_students.iloc[6]
predata_interquartile_75
interquartile = predata_interquartile_75-predata_interquartile_25
interquartile
```

```
box_plot = data.boxplot(column=["freetime","studytime","traveltime"]) #boxplot
box_plot
```

```
plots.scatter(data.studytime,data.freetime)
```

```
fam_size_count = data.groupby("famsize").size() #bar chart
bar_chart = fam_size_count.plot.bar()
bar_chart
```

```
pie_chart = fam_size_count.plot.pie() #pie chart
pie_chart
```

```

histogram_time = data.hist(by='famsize',column=["freetime","studytime"]) #histogram
histogram_time
```
