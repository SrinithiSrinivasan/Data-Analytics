```
df=pd.read_csv("a.csv")
df.columns=['time','tweets']
df = df.head(200)
df
```

```
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
import nltk
nltk.download('punkt')
```

```
#tokenization and stemming
ps = PorterStemmer()
words_stem =[]
column=df['tweets']
for line in column:
  words = word_tokenize(line)
  for w in words:
    words_stem.append(ps.stem(w))
    print(ps.stem(w))
```

```
import re
word_alpha = []
for w in words_stem:
  match = re.search(r'[A-Za-z]+',w)
  if match:
    word_alpha.append(w)
word_alpha
```

```
from nltk import ngrams


threegrams = ngrams(word_alpha, 3)

for grams in threegrams:
  print(grams)
```

```
#document term matrix
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer()
X = vec.fit_transform(column)
df_term = pd.DataFrame(X.toarray(), columns=vec.get_feature_names())
print(df_term)
```
